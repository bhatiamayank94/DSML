# CC

## Dodged countplot

```python
df_top3=df_raw[(df_raw['Publisher'].isin(df_raw['Publisher'].value_counts().index[0:3]))&(df_raw['Platform'].isin(df_raw['Platform'].value_counts().index[0:3]))&(df_raw['Genre'].isin(df_raw['Genre'].value_counts().index[0:3]))]
```

```python
sns.countplot(x='Publisher',hue='Platform',data=df_top3)
```

## Stacked countplot

```python
sns.countplot(x='Publisher',hue='Genre',data=df_top3,dodge=False)
plt.ylim(top=1000,bottom=0)
```

# Continuous categorical

```python
# sns.boxenplot(x='Publisher',data=df_top3,y='Global_Sales')
sns.violinplot(x='Publisher',data=df_top3,y='Global_Sales')
```

# Subplot

```python
plt.subplot(1,3,1)
sns.boxenplot(x='Publisher',data=df_top3,y='Global_Sales')
plt.xticks(rotation=90)


plt.subplot(1,3,3)
sns.violinplot(x='Publisher',data=df_top3,y='Global_Sales')
plt.xticks(rotation=90)
```

```python
fig,ax=plt.subplots(2,2,figsize=(6,4))
ax[0,0].scatter(data=df_top3,x="NA_Sales",y="EU_Sales")
ax[0,1].scatter(data=df_top3,x="NA_Sales",y="JP_Sales")
ax[1,0].scatter(data=df_top3,x="NA_Sales",y="Other_Sales")
ax[1,1].scatter(data=df_top3,x="NA_Sales",y="Global_Sales")
```
