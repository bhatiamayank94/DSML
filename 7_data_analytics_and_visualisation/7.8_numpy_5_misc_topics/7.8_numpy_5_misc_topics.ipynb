# np.split

```python
'''
in the second parameter
    if integer is given - then it will split array into those many equal pieces
    if array is given, values act as indices where split has to be performed
    if array is provided, and index is out of range, then empty array is returned
'''
```

```python
o=np.arange(18).reshape(-1,3)
p=np.arange(6)

print(o)
print()
print(np.split(o,2,0)) # normal split where array is divided for rows into 2 pieces
print()
print(np.split(o,3,1)) # normal split where array is divided for columns into 3 pieces
print()
print(np.split(o,[1,3,7],0)) # array elements act as indices and array is split at indices 1 and 3. note empty array for split at index 7
print()
print(np.split(p,2)) # 1d using equal pieces
print()
print(np.split(p,[1,3,8])) # 1d using indices
print()
print(np.split(p,[1,3,8])) # 1d using using 
print()
# print(np.split(p,[1,3,8],1)) # 1d on column axis - ERROR
```

```python
# hSPLIT AND VSPLIT
q=np.arange(18).reshape(-1,3)
r=np.arange(6)

print(q)
print()
print(r)
print()
print(np.hsplit(q,3)) # vertical
print()
print(np.vsplit(q,2)) # horizaontal
```

# np.stack

```python
s=np.arange(5)
t=np.arange(5).reshape(5,1)
print(s)
print()
print(t)
print()
print(np.vstack([s,s]))
print()
print(np.hstack([s,s])) # note that here dimension is not being changed
print()
print(np.vstack([t,t]))
print()
print(np.hstack([s,s])) # note that here dimension is not being changed
```

# NP.concatenate

```python
# Concat is equivalent to stack only at high level 
# Concatenate doesnt change dimension. hence shape along the axis of concatenation must be same


u=np.arange(3)
v=np.arange(1,4)
w=np.arange(3,1)
x=np.arange(1.)
print(np.concatenate([u,v],0)) # equivalent to hstack
print(np.vstack([u,v]))
print(np.hstack([u,v]))
```

# Image Manipulation

```python
'''
plt.imread
plt.imshow
plt.imread always save as lists within list
np.where
to change values in an array of an image, you have to make a shallow copy
'''


import matplotlib.pyplot as plt
img=plt.imread(r'C:\Users\28962\Desktop\device adoption files\test_3\1002147761_PAINT_COAT_METER.jpg')
```

## manipulate rgb values

```python
plt.imshow(np.where(img>150,255,0))
```

## Manipulating colours

```python
'''
r - 0
g - 1
b - 2
'''
img_2=img.copy()
img_2[0:,:,2]=0
plt.imshow(img_2)
```

## 3d Transpose

```python
img_3=img.copy()
img_3_a=np.transpose(img_3,[1,0,2])
plt.imshow(img_3_a)
```

## Crop image

```python
img_4=img.copy()
img_4_a=img_4[50:150,50:150,:]
plt.imshow(img_4_a)
```

## Plt.imsave

```python
plt.imsave('img_4_a.jpg',img_4_a)
```
