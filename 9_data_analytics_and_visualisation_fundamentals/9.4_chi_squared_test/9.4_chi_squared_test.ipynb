## Example from class - exhibit 3

```python

# heads - observation = 28, expected = 25. Similarly tails,  observation = 22 and expected = 22

test_stat=(((28 - 25)**2) / 25) + (((22 - 25)**2) / 25)
p_value=1-chi2.cdf(test_stat,df=1)

print(p_value)
```

```python
'''
A national survey was conducted to obtain information on the alcohol consumption patterns of U.S. adults by marital status.
A random sample of 1772 residents, aged 18 and older, yielded the data displayed in Table below:

test whether Marital status and alcohol consumption are associated with a 5% significance level.
'''

import pandas as pd

data = {
    "Abstain": [67, 411, 85, 27],
    "1-60":    [213, 633, 51, 60],
    "Over 60": [74, 129, 7, 15]
}
index = ["Single", "Married", "Widowed", "Divorced"]
df = pd.DataFrame(data, index=index)


from scipy.stats import chisquare,chi2_contingency

print(chi2_contingency(df))

```
