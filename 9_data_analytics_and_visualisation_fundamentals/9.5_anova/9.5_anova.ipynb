```python
# q1

data = {
    'Northeast': [13, 8, 11, 12, 11],
    'Midwest':   [15, 10, 16, 11, 13, 10],
    'South':     [5, 11, 9, 5],
    'West':      [8, 10, 6, 5, 7]
}

# Since the columns have different lengths, we need to pad the shorter ones with NaN
from itertools import zip_longest

# Transpose the data to rows, padding with None (which pandas will convert to NaN)
rows = list(zip_longest(
    data['Northeast'],
    data['Midwest'],
    data['South'],
    data['West']
))

df = pd.DataFrame(rows, columns=['Northeast', 'Midwest', 'South', 'West'])

# print(f_oneway(df['Northeast'],df['Midwest'],df['South'],df['West']))  ## This gives error because of nan values
a = [13, 8, 11, 12, 11]
b = [15, 10, 16, 11, 13, 10]
c = [5, 11, 9, 5]
d = [8, 10, 6, 5, 7]

print(f_oneway(a,b,c,d))
```

```python
# q2 
a = [382, 391, 335, 368, 400, 372]
b = [560, 343, 512, 329, 391, 367]
c = [384, 458, 409, 309, 374, 459]
d = [325, 390, 304, 240, 306, 169]
e = [360, 298, 272, 368, 320, 326]

print(f_oneway(a,b,c,d,e))
```
