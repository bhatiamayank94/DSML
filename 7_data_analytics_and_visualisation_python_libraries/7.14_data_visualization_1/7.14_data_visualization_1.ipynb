```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
import seaborn as sns
```

```python
df_raw=pd.read_csv(r'C:\Users\28962\Documents\GitHub\DSML\7_data_analytics_and_visualisation\7.14_data_visualization_1\games.csv')
```

# Basic plot line plot

```python
x_val=list(range(10))
y_val=[math.exp(i) for i in x_val]
plt.plot(x_val,y_val)
```

# Anatomy of charts

## Basic bar chart

```python
genre_count=df_raw['Genre'].value_counts()
x_val=genre_count.index
plt.figure(figsize=(10,2),facecolor="grey")
plt.title("Games per genre",fontsize=20)
plt.xlabel("genre",fontsize=12)
plt.ylabel("# Games",fontsize=12)
plt.xticks(rotation=45,fontsize=7)
plt.yticks(rotation=1,fontsize=7)
plt.ylim(top=max(genre_count)+3000)
for i, value in enumerate(genre_count):
    plt.text(i, value, f'{value:,}', ha="center", va="bottom")
# plt.text(x=10,y=2000,s="hello") # Can be used for data labelling 
plt.bar(x_val,genre_count,width=0.9,align="center",color="cyan")
```

## Pie plot

```python

df_raw_2=df_raw[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df_raw_2.sum(axis=0)
plt.pie(x=df_raw_2.sum(axis=0).values,labels=df_raw_2.sum(axis=0).index,autopct='%1.0f%%')
plt.show()
```

# SNS chart

## countplot - distribution graph

```python
sns.countplot(data=df_raw,x=df_raw['Genre'],order=df_raw['Genre'].value_counts().index,color="cornflowerblue") # Note that value counts are already sorted in descending order 
plt.xticks(rotation=45)
plt.show()
```
