# .apply

## on each cell

```python
def encode(data):
    if data=="Male":
        return 0
    else:
        return 1

df_dir['gender_2']=df_dir['gender'].apply(encode)
```

## on series - note axis is available for .apply() as well

```python
df_mov[["revenue", "budget"]].apply(np.sum)
df_mov[["revenue", "budget"]].apply(np.sum,axis=1)
```

```python
# Profit using apply

def prof(x):
    return x['revenue']-x['budget']

df_mov['profit_2']=df_mov.apply(prof,axis=1)
df_mov['profit']=df_mov[['revenue','budget']].apply(prof,axis=1)
```

# Aggregation

## Value_counts - distribution

```python
df_mer=df_mov.merge(df_dir,how='left',left_on='director_id',right_on="id")
df_mer['director_name'].value_counts()
```

## Count with condition

## Group by 

```python
dir_group=df_mer.groupby('director_name')
```

### Ngroups

```python
dir_group.ngroups
```

### Groups - shows how group works by storing indices

```python
print(dir_group.groups)
```

### aggregations against group by 

```python
# dir_group['title'].count()
dir_group['year'].aggregate(['min','max','count']).sort_values(by='count',ascending=False)
```

### group based filtering

```python
# filter for directors with max greater than 100M

df_mer=df_mov.merge(df_dir,how='left',left_on='director_id',right_on="id")
df_mer['budget']=df_mer['budget']/1000000
df_dir=df_mer.groupby('director_name')['budget'].max().reset_index()
df_dir_max=df_dir[df_dir['budget']>=100]
df_filtered=df_mer[df_mer['director_name'].isin(df_dir_max['director_name'])]

```
