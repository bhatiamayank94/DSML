```python
'''
Classification metrics:
    Accuracy metric: Not good with imbalanced data. For example with a spam vs non spam,even with all non spams, accuracy will be 90%
    Hence, we need to check accuracy and inaccuracies for each class

    Confusion matrix: matrix to track true negative, false positive, false negative, and true positive
        Note - confusion matrix for n classes will be n x n
        false positive is type 1 error and false negative is type 2 error
    
    PRECISION: TP/(TP+FP): Basically, out of predicted class 1, how many were actual right. 
        This is for the cases where marked values should be correct, unmarked can be ignored: 
        examples marking criminals out of a population, marked criminal should definitely be a criminal, '
        unmarked can include criminal as well but innocent shouldnt be marked as criminal
        here threshold will be very high
        No major issues with false negatives

    RECALL/SENSITIVITY: TP/(TP+FN): Basically out of the total actual classes, how many were accurately detected
        For example Detecting cancer. No true cases should be left undetected
        here false negatives should be minimized

    F1: For a model where both recall and precision is required, we use F1 score: Harmonic mean of precision and recall

'''
```

# Code for Classification metrics

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,precision_score,recall_score,f1_score
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\Mayank Bhatia\Documents\GitHub\DSML\13_intro_to_ml_nn\13.10_intro_to_ml_classification_metrics\Spam_finalData.csv')
```

```python
x=df.drop('label_num',axis=1)
y=df['label_num']

x_mid,x_test,y_mid,y_test=train_test_split(x,y)
x_train,x_val,y_train,y_val=train_test_split(x_mid,y_mid)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))

# WE ARE GETTING 93% ACCURACY. IS THIS GOOD? WE DONT KNOW, TILL WE GET CONFUSION MATRIX
```

```python
y_pred_test=model.predict(x_test)
print(model.score(x_test,y_test))
conf_matrix=confusion_matrix(y_test,y_pred_test)
```

```python
ConfusionMatrixDisplay(conf_matrix).plot()
```

```python
print(precision_score(y_test,y_pred_test))
print(recall_score(y_test,y_pred_test))
print(f1_score(y_test,y_pred_test))

# PRECIOSION SCORE RECALL SCORE, BOTH SEEM TO BE GOOD
# EVEN F1 SCORE SEEMS TO BE GOOD
```
