# IQR - Inter quartile range

```python
import pandas as pd
import seaborn as sns
import numpy as np
df=pd.read_csv(r'C:\Users\28962\Documents\GitHub\DSML\8_data_analytics_and_visualisation_probability_and_stats\8.5_descriptive_statistics\weight-height.csv')
```

```python
df.describe(include='all')
sns.boxenplot(data=df['Height'])
```

## Defining IQR

```python
a=df['Height'].quantile(0.25)
b=df['Height'].quantile(0.75)
iqr=b-a
outlier_lower_range=a-1.5*iqr
outlier_upper_range=b+1.5*iqr
print("outlier range is",round(outlier_lower_range,2),"and",round(outlier_upper_range,2)) #Check these numbers with the graph above to see outliers and iqr
print("IQR is ",round(iqr,2),"iqr beginning is ",round(a,2),"iqr end is ",round(b,2)) #Check these numbers with the graph above to see outliers and iqr
```

# CDF 

```python
df.describe(include='all')
x_values=np.linspace(50,80,500)
```

```python
y_values=[]
for i in x_values:
    y_temp=df[df['Height']<i].shape[0]
    tot=df.shape[0]
    prob=y_temp/tot
    y_values.append(prob)
```

```python
sns.lineplot(y=y_values,x=x_values)
```
