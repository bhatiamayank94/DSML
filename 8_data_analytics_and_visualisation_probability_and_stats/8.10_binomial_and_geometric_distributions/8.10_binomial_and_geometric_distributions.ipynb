# Binomial from scipy

```python
'''
scipy.stats.binom.pmf(k, n, p)
k: Number of successes (can be a single value or an array of values)
n: Number of trials (must be an integer, 
p: Probability of success on each trial (between 0 and 1)

'''
```

```python
from scipy.stats import binom, geom
import numpy as np 
import seaborn as sns 
```

```python
# probability of 0,1,2,3 interview successes 3 interviews
binom.pmf([0,1,2,3],3,0.1)
```

```python
# Coin toss probability 

prob = binom.pmf(np.arange(0,16),15,0.5)
sns.barplot(x=np.arange(0,16),y=prob)
```

```python
# You have 10 interviews, what is the probability that you would succeed in at most 5 interviews, given that probability of success is 0.2?

prob=binom.pmf(np.arange(0,6),10,0.2)
print(prob.sum())
```

# Geometric from scipy

## geom.pmf

```python
# coin toss case

prob=geom.pmf(np.arange(1,11),0.5)
sns.barplot(x=np.arange(1,11),y=prob)
```

## geom.cdf

```python
prob=geom.cdf(3,0.5)
print(prob)
```

```python
# for a 5% chance of winning, you will most probably win in 80th+ iteration
sns.lineplot(x=np.arange(1,101),y=geom.cdf(np.arange(1,101),0.05))
```
