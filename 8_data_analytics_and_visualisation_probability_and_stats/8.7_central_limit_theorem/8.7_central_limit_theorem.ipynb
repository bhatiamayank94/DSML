```python
'''
CLT - Central Limit Theorem

Average of bunch of samples out of random distributions always follow normal distribution even if original distribution is not normal 

Forumalae good to know
    E[x']=mu (Total average)
    SD[x']=SD/sqrt(n)

With increase in sample size, SD of the samples mean decreases ( because of sqrt of SD) and hence spread decreases and 
hence normal distribution becoms thinner with increasing n

'''
```

# Testing of CLT

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python

a=np.random.randint(100000,size=(10000))
print(a)
# sns.histplot(a) # CLearly this data is not normally distributed

def clt (iters, sample_size ):
    means=[]
    for i in range(iters):
        k=np.mean(random.sample(list(a),sample_size))
        means.append(k)
    return means
    
means = clt(1000,2) # try this with different combo of parameters. This will always hold true
sns.histplot(means) 
# sns.histplot(a)
```
